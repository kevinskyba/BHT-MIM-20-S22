```python
import pandas as pd

df = pd.read_csv("edlich-kmeans-A0.csv")
df.describe()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(df['V1'], df['V2'], df['V3'], cmap=plt.cm.viridis, linewidth=0.2)
plt.show()
```

# A) What is the best k? (usually derived by plotting each graph ;-)


The best k is 5, or possibly 4 depending on the domain, since the right cluster might be considered one.

# B) Can you print the clustering vector?



```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
kmeans = kmeans.fit(df)

labels = kmeans.predict(df)
C = kmeans.cluster_centers_
C
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(df['V1'], df['V2'], df['V3'], cmap=plt.cm.viridis, linewidth=0.2)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker="*", s=1000)
plt.show()
```
